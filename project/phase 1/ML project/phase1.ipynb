### (C) Visualization of data

```python
#plotting usage of each color on a scatter plot
df=pd.DataFrame(data,columns=['index','R','G','B'])
plt.scatter(df['index'],df['R'],color='red')
plt.title('color vs Red intensity')
plt.xlabel('color')
plt.ylabel('red')
plt.show()
plt.scatter(df['index'],df['G'],color='green')
plt.title('color vs Green intensity')
plt.xlabel('color')
plt.ylabel('green')
plt.show()
plt.scatter(df['index'],df['B'],color='blue')
plt.title('color vs Blue intensity')
plt.xlabel('color')
plt.ylabel('blue')
plt.show()
```

```python
#intensity of each color vs color name
x=np.array(df['R'])
plt.hist(x,color='red')
plt.show() 
x=np.array(df['G'])
plt.hist(x,color='green')
plt.show()
x=np.array(df['B'])
plt.hist(x,color='blue')
plt.show() 
```

```python
#data cleaning and preparation
column_names=df.columns
print(column_names)
df.dtypes
#redundancy check
for i in column_names:
    print('{} is unique: {}'.format(i, df[i].is_unique))
# Check the index values
df.index.values
# Check if a certain index exists
300 in df.index.values
df['R'] = df['R'].fillna(' ')
df['G'] = df['G'].fillna(' ')
df['B'] = df['B'].fillna(' ')
```
