```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,1,2,3,4,3,4,6,4])
def h_x(x):
    return 1.2+.5*x
plt.plot(x,h_x(x))
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('first graph')
plt.show()
```

```python
A=np.array([1,1,2,3,4,3,4,6,4])
B=np.array([2,1,0.5,1,3,3,2,5,4])
dot=np.dot(A,B)
print(dot)
```

```python
plt.plot(A,B)
plt.xlabel('A')
plt.ylabel('B')
plt.title('second graph')
plt.show()
```

```python
MSE=np.sum(np.square(A-B))/A.shape
print(MSE)
```

```python
x=np.array([1,1,2,3,4,3,4,6,4])
def h_x(x):
    return .8+.5*x
for i in x:
    MSE=np.sum(np.square(h_x(i)-i))/x.shape
    MSE=MSE/2
print(MSE)
```
